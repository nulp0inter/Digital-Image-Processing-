```python
# Slide 3
# Image Data
# https://www.imageprocessingplace.com/DIP-3E/dip3e_book_images_downloads.htm

import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_img_and_hist(img, title_img, title_hist, is_histogram=False):
    plt.figure(figsize=(12, 5))

    # Left: Image
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title(title_img)
    plt.axis('off')

    # Right: Histogram or Transformed Image
    plt.subplot(1, 2, 2)
    if is_histogram:
        # Calculate histogram from scratch using numpy
        counts, bins = np.histogram(img.flatten(), bins=256, range=[0,256])
        plt.bar(bins[:-1], counts, width=1, color='black')
        plt.xlim([0, 255])
        plt.title(title_hist)
    else:
        plt.imshow(title_hist, cmap='gray', vmin=0, vmax=255) # title_hist is actually the image here
        plt.title("Transformed Image")
        plt.axis('off')

    plt.show()
```

1. Plot Image Histograms

```python
# 3.1.1 – Dark Image (Left-skewed histogram)
# Load the image
image_path = 'DIP3E_Original_Images_CH03/Fig0308(a)(fractured_spine).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plot_img_and_hist(img, "Dark Image", "Left-Skewed Histogram", is_histogram=True)
```

2. Contrast Stretching:
Linear transformations to expand the range of intensity levels.

```python
# 3.2.1 – Full Dynamic Range Stretching (Normalization to 0-255)
# Load the image
image_path = 'DIP3E_Original_Images_CH03/Fig0310(b)(washed_out_pollen_image).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: out = (in - min) * (255 / (max - min))
min_val = np.min(img)
max_val = np.max(img)
stretched = (img - min_val) * (255.0 / (max_val - min_val))
stretched = np.uint8(stretched)

plot_img_and_hist(img, "Original Low Contrast", stretched, is_histogram=False)
```

```python
# 3.2.2 – Dark-End Stretching (Shadow Detail Recovery)
# Load the image
image_path = 'DIP3E_Original_Images_CH03/Fig0308(a)(fractured_spine).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: Slope > 1 for lower values (r < 100), Slope < 1 for higher values
r1, s1 = 100, 180  # Push input 100 up to output 180 (brightening shadows)
r2, s2 = 255, 255

# Apply piecewise function
stretched = np.zeros_like(img, dtype=np.float32)
mask1 = img < r1
mask2 = img >= r1
stretched[mask1] = (s1 / r1) * img[mask1]
stretched[mask2] = ((s2 - s1) / (r2 - r1)) * (img[mask2] - r1) + s1
stretched = np.uint8(stretched)

plot_img_and_hist(img, "Original Dark", stretched, is_histogram=False)
```

```python
# 3.2.3 – Saturated (Clipping) Stretching (ignoring outliers)
# Load the image
image_path = 'DIP3E_Original_Images_CH03/Fig0310(b)(washed_out_pollen_image).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: Clip bottom 5% and top 5% of range to force contrast in the middle
low_clip, high_clip = 100, 160  # Manual selection based on histogram
stretched = (img.astype(float) - low_clip) * (255.0 / (high_clip - low_clip))
stretched = np.clip(stretched, 0, 255).astype(np.uint8)

plot_img_and_hist(img, "Original", stretched, is_histogram=False)
```

3. Piecewise Linear Transformation:
General form implementation.

```python
# 3.3 – General Piecewise Linear Transformation (Contrast Stretching with r1,s1, r2,s2)
# Load the image
image_path = 'DIP3E_Original_Images_CH02/Fig0241(a)(einstein low contrast).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: Define points (r1, s1) and (r2, s2) to shape the contrast curve
r1, s1 = 70, 20
r2, s2 = 150, 230
output = np.zeros_like(img, dtype=np.float32)

# Region 1: 0 to r1
mask1 = img < r1
output[mask1] = (s1 / r1) * img[mask1]
# Region 2: r1 to r2 (High Slope = High Contrast)
mask2 = (img >= r1) & (img < r2)
output[mask2] = ((s2 - s1) / (r2 - r1)) * (img[mask2] - r1) + s1
# Region 3: r2 to 255
mask3 = img >= r2
output[mask3] = ((255 - s2) / (255 - r2)) * (img[mask3] - r2) + s2

output = np.uint8(output)
plot_img_and_hist(img, "Original", output, is_histogram=False)
```

4. Histogram Equalization & Transfer Functions

```python
# 3.4.1 – Global Histogram Equalization (Uniform Distribution)
# Load the image
image_path = 'DIP3E_Original_Images_CH02/Fig0241(a)(einstein low contrast).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: Calculate PDF -> CDF -> Map to 0-255
hist, bins = np.histogram(img.flatten(), 256, [0, 256])
cdf = hist.cumsum()
cdf_normalized = cdf * 255 / cdf[-1] # Normalization
equalized = np.interp(img.flatten(), bins[:-1], cdf_normalized).reshape(img.shape)
equalized = np.uint8(equalized)

plot_img_and_hist(img, "Original", equalized, is_histogram=False)
```

```python
# 3.4.2 – Logarithmic Transfer Function (Expands dark pixels)
# Load the image (Spectrum is ideal for log)
image_path = 'DIP3E_Original_Images_CH03/Fig0305(a)(DFT_no_log).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: s = c * log(1 + r)
c = 255 / np.log(1 + np.max(img))
log_img = c * np.log(1 + img.astype(np.float32))
log_img = np.uint8(log_img)

plot_img_and_hist(img, "Original Spectrum", log_img, is_histogram=False)
```

```python
# 3.4.3 – Power-Law (Gamma) Transfer Function
# Load the image
image_path = 'DIP3E_Original_Images_CH03/Fig0308(a)(fractured_spine).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: s = c * r^gamma
gamma = 0.5 # Gamma < 1 for dark image enhancement
img_norm = img / 255.0
gamma_img = np.power(img_norm, gamma) * 255.0
gamma_img = np.uint8(gamma_img)

plot_img_and_hist(img, "Original Dark", gamma_img, is_histogram=False)
```

```python
# 3.4.4 – Sigmoid (S-Curve) Transfer Function (Mid-tone Contrast Boost)
# Load the image
image_path = 'DIP3E_Original_Images_CH02/Fig0241(a)(einstein low contrast).tif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Logic: s = 1 / (1 + exp(-gain * (r - cutoff)))
# This pushes values away from the center (cutoff)
gain = 10
cutoff = 0.5
img_norm = img.astype(float) / 255.0
sigmoid_img = 1 / (1 + np.exp(-gain * (img_norm - cutoff)))
sigmoid_img = (sigmoid_img * 255).astype(np.uint8)

plot_img_and_hist(img, "Original", sigmoid_img, is_histogram=False)
```
